```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([1, 2])
b = np.array([3, 1])
a_plus_b = a + b

plt.figure(figsize=(7, 7))

plt.arrow(0, 0, a[0], a[1], head_width=0.05, head_length=0.075, length_includes_head=True, fc='orange', ec='orange')
plt.arrow(0, 0, b[0], b[1], head_width=0.05, head_length=0.075, length_includes_head=True, fc='orange', ec='orange')
plt.arrow(0, 0, a_plus_b[0], a_plus_b[1], head_width=0.05, head_length=0.075, length_includes_head=True, fc='orange', ec='orange')

mid_a = a / 2
mid_b = b / 2
mid_a_plus_b = a_plus_b / 2

plt.annotate(r'$(a_1,a_2)$', (a[0], a[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)
plt.annotate(r'$(b_1,b_2)$', (b[0], b[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)
plt.annotate(r'$(a_1+b_1,a_2+b_2)$', (a_plus_b[0], a_plus_b[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)

plt.annotate(r'$\mathbf{a}$', (mid_a[0], mid_a[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)
plt.annotate(r'$\mathbf{b}$', (mid_b[0], mid_b[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)
plt.annotate(r'$\mathbf{a+b}$', (mid_a_plus_b[0], mid_a_plus_b[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)

plt.plot([a[0], a_plus_b[0]], [a[1], a_plus_b[1]], linestyle='--', lw=1, color='white') 
plt.plot([b[0], a_plus_b[0]], [b[1], a_plus_b[1]], linestyle='--', lw=1, color='white')

par_b_a_b2 = np.array([a[0] + b[0], b[1]])
plt.plot([b[0], par_b_a_b2[0]], [b[1], par_b_a_b2[1]], linestyle='--', lw=1, color='white')
plt.plot([par_b_a_b2[0], a_plus_b[0]], [par_b_a_b2[1], a_plus_b[1]], linestyle='--', lw=1, color='white')

plt.annotate(r'$(a_1+b_1, b_2)$', (par_b_a_b2[0], par_b_a_b2[1]), textcoords="offset points", xytext=(10,10), ha='center', fontsize=12)

plt.axhline(0, color='white',linewidth=0.5)
plt.axvline(0, color='white',linewidth=0.5)
plt.title('Parallelogram Law')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([1, 2])
t = 2
a_mult_t = a * t
plt.figure(figsize=(7, 7))

plt.arrow(0, 0, a[0], a[1], head_width=0.05, head_length=0.075, length_includes_head=True, fc='orange', ec='orange')
plt.arrow(0, 0, a_mult_t[0], a_mult_t[1], head_width=0.05, head_length=0.075, length_includes_head=True, fc='orange', ec='orange')

mid_a = a / 2
mid_t = a_mult_t * 0.75
# mid_a_plus_b = a_plus_b / 2

plt.annotate(r'$(a_1,a_2)$', (a[0], a[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)
plt.annotate(r'$\mathbf{a}$', (mid_a[0], mid_a[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)
plt.annotate(r'$t\mathbf{a}$', (mid_t[0], mid_t[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)
plt.annotate(r'$(ta_1,ta_2)$', (a_mult_t[0], a_mult_t[1]), textcoords="offset points", xytext=(10,5), ha='center', fontsize=14)


plt.axhline(0, color='white',linewidth=0.5)
plt.axvline(0, color='white',linewidth=0.5)
plt.title('Scalar Multiplication')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.show()
```
